```python
import pandas as pd
import re
```

```python
monster_data = pd.read_csv('MONSTER.csv',usecols=[8,11,12])
monster_data
```

```python
s1 = monster_data.dropna(how='any')['location']
s2 = monster_data.dropna(how='any')['salary']
s3 = monster_data.dropna(how='any')['sector']
```

```python
location = s1.str.split(',',expand=True)[1]
location
```

```python
first = pd.DataFrame({'location': location,
                      'salary': s2,
                      'sector': s3    
})
first
```

```python
pattern = '[\$]'
#second = []
first.s1 = []
first.s2 = []
first.s3 = []
for n in range(0,2611):
    kk=first.iloc[n,0:3]
    regex = kk.str.contains(pattern)
    if regex[1] == True:
        first.s1.append(kk[0])
        first.s2.append(kk[1])
        first.s3.append(kk[2])
second = pd.DataFrame({'location': first.s1,
                      'salary': first.s2,
                      'sector': first.s3    
})
second        
```

```python
pattern = '[-]'
#second = []
first.s4 = []
first.s5 = []
first.s6 = []
for n in range(0,2214):
    kk=second.iloc[n,0:3]
    regex = kk.str.contains(pattern)
    if regex[1] == True:
        first.s4.append(kk[0])
        first.s5.append(kk[1])
        first.s6.append(kk[2])
second_1 = pd.DataFrame({'location': first.s4,
                      'salary': first.s5,
                      'sector': first.s6    
})
second_1
```

```python
rule=re.compile(r'[^a-zA-z]')
#second = []
first.s4 = []
first.s5 = []
first.s6 = []
for n in range(0,1993):
    kk=second.iloc[n,0:3]
    uu = str(kk[0])
    result = rule.sub(' ',uu).strip()
    if len(result)>2:
        result = ''
    first.s4.append(result)
    first.s5.append(kk[1])
    first.s6.append(kk[2])
second_1 = pd.DataFrame({'location': first.s4,
                      'salary': first.s5,
                      'sector': first.s6    
})
second_1
```
